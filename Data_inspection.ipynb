```
import import_ipynb
import Utils as ut
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```
images, labels = ut.load_training()
```

```
frequency = ut.frequency(labels)
classes = ['baseline', 'mass_b', 'mass_m', 'calc_b', 'calc_m']
colors = ['tab:blue', 'tab:green', 'tab:olive', 'tab:red', 'tab:orange']
plt.bar(classes, frequency, width=0.6, align='center', color=colors)
plt.xlabel('classes')
plt.ylabel('frequency')
plt.title('Class distibution original dataset', fontdict=dict(fontsize=14, fontweight='bold'))
plt.show()
```

# Abnormality classification

```
mc_set, mc_lab = ut.binarize(images, labels, 'mass_calcification')
```

```
mass_sample = mc_set[:5]
calcification_sample = mc_set[-5:]

ut.plot_sample(mass_sample, 'Mass')
ut.plot_sample(calcification_sample, 'Calcification')
```

```
frequency_mc = ut.frequency(mc_lab)
classes_mc = ['mass', 'calcification']
colors_mc = ['tab:blue', 'tab:cyan']

plt.bar(classes_mc, frequency_mc, width=0.8, align='center', color=colors_mc)
plt.xlabel('classes')
plt.ylabel('frequency')
plt.title('Class distibution', fontdict=dict(fontsize=14, fontweight='bold'))
plt.show()
```

# Abnormality diagnosis

```
bm_set, bm_lab = ut.binarize(images, labels, 'benign_malignant')
```

```
banign_sample = bm_set[:5]
malignant_sample = bm_set[-5:]
ut.plot_sample(banign_sample, 'Benign')
ut.plot_sample(malignant_sample, 'Malignant')
```

```
frequency_mc = ut.frequency(bm_lab)
classes_mc = ['benign', 'malignant']
colors_mc = ['tab:orange', 'tab:red']

plt.bar(classes_mc, frequency_mc, width=0.8, align='center', color=colors_mc)
plt.xlabel('classes')
plt.ylabel('frequency')
plt.title('Class distibution', fontdict=dict(fontsize=14, fontweight='bold'))
plt.show()
```
